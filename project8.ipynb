```python
import numpy as np
```

```python
class DataAnalytics:

    def __init__(self):
        self.array = None

    @staticmethod
    def welcome_message():
        print("    NumPy Analyzer - Final Project")

    def _check_array(self):
        if self.array is None:
            print("\n You must create an array first (choose Option 1).")
            return False
        return True



    def create_array(self):
        print("\nCreate Array")
        print("1. 1D Array")
        print("2. 2D Array")
        print("3. 3D Array")

        try:
            choice = int(input("Enter choice (1-3): "))
            match choice:
                case 1:
                    size = int(input("How many numbers?: "))
                    elements = list(map(int, input(f"Enter {size} elements separated by space: ").split()))
                    self.array = np.array(elements)
                    print("Array Created:\n", self.array)
                case 2:
                    rows = int(input("Rows: "))
                    cols = int(input("Columns: "))
                    elements = list(map(int, input(f"Enter {rows*cols} elements separated by space: ").split()))
                    self.array = np.array(elements).reshape(rows, cols)
                    print("Array Created:\n", self.array)
                case 3:
                    depth = int(input("Depth: "))
                    rows = int(input("Rows: "))
                    cols = int(input("Columns: "))
                    elements = list(map(int, input(f"Enter {depth*rows*cols} elements separated by space: ").split()))
                    self.array = np.array(elements).reshape(depth, rows, cols)
                    print("Array Created:\n", self.array)
                case _:
                    print("Invalid choice")
        except Exception as e:
            print(f"Input Error: {e}")



    def indexing_slicing(self):
        if not self._check_array(): return

        print("\nIndexing & Slicing")
        print("1. Indexing")
        print("2. Slicing")
        print("3. Go Back")

        try:
            choice = int(input("Enter choice: "))
            flat_arr = self.array.flatten()

            match choice:
                case 1:
                    index = int(input(f"Enter index (0 to {flat_arr.size - 1}): "))
                    if 0 <= index < flat_arr.size:
                        print("Value:", flat_arr[index])
                    else:
                        print("Index out of bounds")

                case 2:
                    start = int(input("Start index: "))
                    end = int(input("End index: "))
                    print("Slice:", flat_arr[start:end])

                case 3:
                    return

                case _:
                    print("Invalid choice")
        except ValueError:
            print("Invalid input.")



    def mathematical_operations(self):
        if not self._check_array(): return

        print("\nMath Operations")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")


        try:
            choice = int(input("Enter choice: "))

            print(f"Enter the same-size array elements ({self.array.size} elements separated by space):")
            input_str = input()
            elements = list(map(int, input_str.split()))

            if len(elements) != self.array.size:
                print(f"Error: You must enter exactly {self.array.size} numbers.")
                return

            other = np.array(elements).reshape(self.array.shape)

            print("\nOriginal Array:\n", self.array)
            print("Second Array:\n", other)

            match choice:
                case 1: print("Result of Addition:\n", self.array + other)

                case 2: print("Result of Subtraction:\n", self.array - other)

                case 3: print("Result of Multiplication:\n", self.array * other)

                case 4: print("Result of Division:\n", self.array / other)


                case _:
                    print("Invalid choice")

        except ValueError:
            print("Please enter valid numbers.")
        except Exception as e:
            print(f"Error: {e}")



    def combine_split(self):
        if not self._check_array(): return

        print("\nCombine & Split")
        print("1. Combine")
        print("2. Split")

        try:
            choice = int(input("Enter choice: "))
            match choice:
                case 1:
                    print(f"Enter elements to combine ({self.array.size} elements):")
                    elements = list(map(int, input().split()))
                    other = np.array(elements).reshape(self.array.shape)

                    print("Original:\n", self.array)

                    print("Second:\n", other)

                    print("Combined (Vertical Stack):\n", np.vstack((self.array, other)))

                case 2:
                    parts = int(input("How many parts?: "))
                    result = np.array_split(self.array, parts)
                    for i, arr in enumerate(result):
                        print(f"Part {i+1}:\n{arr}")
                case _:
                    print("Invalid choice")
        except Exception:
            print("Error in input or split.")



    def search_sort_filter(self):
        if not self._check_array(): return

        print("\nSearch, Sort, Filter")
        print("1. Search")
        print("2. Sort")
        print("3. Filter")

        try:
            choice = int(input("Enter choice: "))
            match choice:
                case 1:
                    val = int(input("Value to search: "))
                    print("Indices:", np.where(self.array == val))

                case 2:
                    print("Sorted:\n", np.sort(self.array, axis=None))

                case 3:
                    val = int(input("Show numbers greater than: "))
                    print("Filtered:", self.array[self.array > val])

                case _:
                    print("Invalid choice")

        except ValueError:
            print("Invalid input.")

    def aggregates(self):
        if not self._check_array(): return

        print("\nStatistics")
        print("1. Sum")
        print("2. Mean")
        print("3. Median")
        print("4. Std Dev")
        print("5. Variance")


        try:
            choice = int(input("Enter choice: "))
            match choice:
                case 1: print("Sum:", np.sum(self.array))

                case 2: print("Mean:", np.mean(self.array))

                case 3: print("Median:", np.median(self.array))

                case 4: print("Std:", np.std(self.array))

                case 5: print("Variance:", np.var(self.array))

                case _: print("Invalid choice")
        except ValueError:
            print("Invalid input.")


def main():
    DataAnalytics.welcome_message()
    obj = DataAnalytics()

    while True:
        print("\n" + "-"*30)
        print("1. Create Array")
        print("2. Indexing & Slicing")
        print("3. Math Operations")
        print("4. Combine / Split")
        print("5. Search / Sort / Filter")
        print("6. Aggregates & Stats")
        print("7. Exit")
        print("-" * 30)

        try:
            choice = int(input("Enter choice: "))
            match choice:
                case 1:
                    obj.create_array()
                case 2:
                    obj.indexing_slicing()
                case 3:
                    obj.mathematical_operations()
                case 4:
                    obj.combine_split()
                case 5:
                    obj.search_sort_filter()
                case 6:
                    obj.aggregates()
                case 7:
                    print("Exiting... Goodbye!")
                    break
                case _:
                    print("Invalid choice")
        except ValueError:
            print("Please enter a valid number.")

if __name__ == "__main__":
    main()
```
